## Weather API


The weather forecast of the arrival city will be obtained from the free API weatherbit.io.

```python
#First define the keys route:
def env_route(key):
    dotenv_path = join(dirname("../.env"), '.env')
    load_dotenv(dotenv_path)
    weather_api_key = os.environ.get(key)
    return weather_api_key
```

```python
#Get key
api_weather_key = env_route('api_key')
print(weather_api_key)
```

```python
# Let's take an example:
arrival_city = 'Rome'
arrival_country_code = 'IT'
```

```python
# Get information from API
url = f'https://api.weatherbit.io/v2.0/forecast/daily?city={arrival_city}&country={arrival_country_code}&key={api_weather_key}'
html_weather = requests.get(url).json()
html_dic = html_weather['data']
html_dic
```

```python
weather_df = pd.DataFrame(html_dic)
weather_df
```

```python
def date_short(day, month, year):  # To put input date in useful format (text)
    short_month = month[0:3]
    date_format = f'{day} {short_month} {year}'
    return date_format

```

```python
date = date_short(10, 'May', 2021)
```

```python
def date_for_weather(date): # Puts the date in format readable by weather_df
    date_2 = datetime.strptime(date, "%d %b %Y").isoformat()
    date_changed = date_2[:10]
    return date_changed
```

```python
date_weather = date_for_weather(date)
date_weather
```

```python
arrival_weather_day_df = weather_df.loc[weather_df['datetime'] == date_weather]
arrival_weather_day_df
```
